```python
from datetime import datetime
import pandas as pd
from forex_python.converter import CurrencyRates

# Tarihi al
today = datetime.now().strftime('%Y-%m-%d')

# forex-python ile döviz kurlarını çek
c = CurrencyRates()
usd_try = c.get_rate('USD', 'TRY')  # Dolar/TL
eur_try = c.get_rate('EUR', 'TRY')  # Euro/TL
eur_usd = c.get_rate('EUR', 'USD')  # Euro/Dolar

# Örnek olarak popüler endekslerin yakın tahmini değerlerini manuel ekliyoruz
# Çünkü forex-python endeks verisi sağlamaz. YFinance gibi bir kaynakla entegre edilebilir.
bist100_close = 8500  # Örnek BIST 100 kapanış değeri
sp500_close = 4500    # Örnek S&P 500 kapanış değeri
nasdaq_close = 15000  # Örnek Nasdaq kapanış değeri

# Verileri birleştir
results = {
    "Date": today,
    "BIST 100 (Kapanış)": bist100_close,
    "S&P 500 (Kapanış)": sp500_close,
    "Nasdaq (Kapanış)": nasdaq_close,
    "USD/TRY (Anlık)": usd_try,
    "EUR/TRY (Anlık)": eur_try,
    "EUR/USD (Anlık)": eur_usd,
}

# Sonuçları bir DataFrame'e dönüştür
df = pd.DataFrame([results])

# Excel dosyasına kaydet
excel_file_path = "market_data.xlsx"
df.to_excel(excel_file_path, index=False)

print(f"Veriler Excel dosyasına kaydedildi: {excel_file_path}")
```
